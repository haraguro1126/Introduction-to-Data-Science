```
# -*- coding: utf-8 -*-
"""IDS_20200529_4.ipynb
Automatically generated by Colaboratory.
Original file is located at
    https://colab.research.google.com/drive/1qLRhpW-pc1o9CjpNXUJ-fR-1jrr9FCtf
"""

x = list(range(1, 101))
N = len(x)
x_bar = sum(x) / N
sse = 0
for xi in x:
    #error = xi - x_bar
    squared_error = (xi - x_bar)**2
    sse += squared_error #sse = sse + squared_error
sample_mse = sse / (N-1)
sample_stdev = sample_mse**(0.5)
print(sample_stdev)
```

```
# -*- coding: utf-8 -*-
"""IDS_20200529_11.ipynb
Automatically generated by Colaboratory.
Original file is located at
    https://colab.research.google.com/drive/1IqnkUaKhkBHgfPdImiNFE4BDH2uB-hIK
"""

def is_prime(x):
    """
    Returns True if x is a prime, or returns False
    """
    divisors = []
    for i in range(1, x+1):
        if x % i == 0:
            divisors.append(i)
    n_divisors = len(divisors)
    return n_divisors == 2
```
